```python
df = pd.read_csv('/kaggle/input/body-fat-extended-dataset/BodyFat - Extended.csv')
```

```python
# viewing the data frame
df
```

```python
# Example: Histogram of body fat distribution
plt.hist(df['BodyFat'], bins=20)
plt.xlabel('Body Fat %')
plt.ylabel('Frequency')
plt.title('Distribution of Body Fat')
plt.show()
```

```python
#describing age and sex distribution
fig, axs = plt.subplots(1, 2, figsize=(25, 10))

# row of sub-plots, one for sex and one for age
axs[0].hist(df['Age'], bins=30, color='blue', alpha=0.7)
axs[0].set_title('Age distribution')

axs[1].hist(df['Sex'], bins=30, color='red', alpha=0.7)
axs[1].set_title('Sex distribution')

plt.show()
```

```python
#describing weight and height distribution
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

# row of sub-plots, one for sex and one for age
axs[0].hist(df['Weight'], bins=30, color='blue', alpha=0.7)
axs[0].set_title('Weight distribution')

axs[1].hist(df['Height'], bins=30, color='green', alpha=0.7)
axs[1].set_title('Height distribution')

plt.show()
```

```python
# describing the rest of the body parts in distribution
fig, axs = plt.subplots(5, 2, figsize=(15, 10))

# row of sub-plots, one for sex and one for age
axs[0, 0].hist(df['Neck'], bins=30, color='blue', alpha=0.7)
axs[0, 0].set_title('Neck distribution')

axs[0, 1].hist(df['Chest'], bins=30, color='green', alpha=0.7)
axs[0, 1].set_title('Chest distribution')

axs[1, 0].hist(df['Abdomen'], bins=30, color='blue', alpha=0.7)
axs[1, 0].set_title('Abs distribution')

axs[1, 1].hist(df['Hip'], bins=30, color='green', alpha=0.7)
axs[1, 1].set_title('Hips distribution')

axs[2, 0].hist(df['Thigh'], bins=30, color='blue', alpha=0.7)
axs[2, 0].set_title('Thighs distribution')

axs[2, 1].hist(df['Knee'], bins=30, color='green', alpha=0.7)
axs[2, 1].set_title('Knees distribution')

axs[3, 0].hist(df['Ankle'], bins=30, color='blue', alpha=0.7)
axs[3, 0].set_title('Ankles distribution')

axs[3, 1].hist(df['Biceps'], bins=30, color='green', alpha=0.7)
axs[3, 1].set_title('Biceps distribution')

axs[4, 0].hist(df['Forearm'], bins=30, color='blue', alpha=0.7)
axs[4, 0].set_title('Forearms distribution')

axs[4, 1].hist(df['Wrist'], bins=30, color='green', alpha=0.7)
axs[4, 1].set_title('Wrists distribution')

plt.subplots_adjust(hspace=0.5)

plt.show()
```
